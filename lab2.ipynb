Згенерувати випадкові дані з np.random для 6 графіків з різними
значеннями у різних проміжках, конвертувати дані в DataFrame,
зчитати та відобразити дані.

```python
import pandas as pd
import numpy as np
import plotly.express as px

x = np.arange(1, 101)

y1 = np.random.uniform(low = 20, high = 60, size=(1,100)).flatten()
y2 = np.random.uniform(low = 30, high = 90, size=(1,100)).flatten()
y3 = np.random.uniform(low = 100, high = 500, size=(1,100)).flatten()
y4 = np.random.uniform(low = 1000, high = 3000, size=(1,100)).flatten()

df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2, 'y3': y3, 'y4': y4})
df

fig = px.line(df, x="x", y=["y1", 'y2', 'y3',	'y4'])
fig.show()
```

Нормалізуавати дані за допомогою MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['y1', 'y2', 'y3', 'y4']] = scaler.fit_transform(df[['y1', 'y2', 'y3', 'y4']])

fig = px.line(df, x="x", y=["y1", 'y2', 'y3', 'y4'])
fig.show()
```

Згенерувати 2 розподіли з різними значеннями проміжку за
допомогою np.random.normal.

```python
import numpy as np
import plotly.express as px

value1 = 10
std_dev1 = 2
size1 = 1000

distribution1 = np.random.normal(value1, std_dev1, size1)

value2 = 20
std_dev2 = 3
size2 = 1000

distribution2 = np.random.normal(value2, std_dev2, size2)

data = {'Distribution 1': distribution1, 'Distribution 2': distribution2}
df = pd.DataFrame(data)

fig = px.histogram(df, x=["Distribution 1", "Distribution 2"], opacity=0.75, barmode='overlay')
fig.show()
```

Стандартизувати попередньо згенеровані розподіли за
допомогою StandardScaler та відобразити на графіку.

```python
from sklearn.preprocessing import StandardScaler

data = {'Distribution 1': distribution1, 'Distribution 2': distribution2}
df = pd.DataFrame(data)

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

fig = px.histogram(df_standardized, x=["Distribution 1", "Distribution 2"], opacity=0.75, barmode='overlay')
fig.show()
```

Використовуючи модель з лабораторної роботи 1 (про реагування
персонажів) потрібно використати оригінальні дані з файлу
alarms_no_scaled.csv, навчити та протестувати модель на цих
даних.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models

# Зчитуємо дані
df = pd.read_csv('alarms_no_scalse.csv')

# Розділяємо дані на ознаки (x) та мітки (y)
x = df[['sound', 'distance', 'visibility']].values
y = df[['alarm']].values

# Розділяємо дані на тренувальний та тестувальний набори
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Створюємо модель
model = models.Sequential([
    layers.Dense(1, input_shape=(x_train.shape[1],), activation='sigmoid')
])

# Компілюємо модель
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Навчаємо модель
model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))

# Оцінюємо модель
loss, accuracy = model.evaluate(x_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)




```
